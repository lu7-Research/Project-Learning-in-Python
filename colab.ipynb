Projeto do Google Research, um ambiente "gratuito" baseado em Jupyter que nos permite criar Jupyter notebooks para escrever e executar códigos Python, R, JS, Julia e outros, além de frameworks de aprendizado de máquina , como Pandas , PyTorch, Tensorflow, Keras, Monk, OpenCV e outros) em um navegador da web.<br><br>
É um shell ou kernel na forma de um processador de texto, onde podemos escrever e executar código. Os dados necessários para o processamento no Google Colab podem ser montados no Google Drive ou importados de qualquer fonte na internet. OPEN-SOURCE.<br><br>
O Google Colab não requer configuração para começar e fornece acesso "gratuito" às GPUs. Uma das principais funcionalidades do Google Colab é que ele permite que qualquer pessoa compartilhe código ativo, equações matemáticas, visualizações de dados, processamento de dados (limpeza e transformação), simulações numéricas, modelos de aprendizado de máquina e muitos outros projetos com outras pessoas.

#Por que usar?
O Google Colab possui recursos exclusivos e essenciais:<br>
1. Ele fornece um ambiente Jupyter notebook gratuito. 
2. Ele vem com pacotes pré-instalados, focados em DATA SCIENCE.
3. Ele é hospedado inteiramente no Google Cloud.
4. Os usuários não precisam fazer a configuração em servidores ou estações de trabalho.
5. Notebooks são salvos automaticamente no Google Drive.
6. Ele fornece Jupyter notebooks em seu navegador.
7. É "gratuito" e oferece GPU e TPU(para projetos maiores terá de usar o Colab Pro).
8. Ele fornece dois aceleradores de hardware:
* GPU (Unidade de processamento gráfico).
* TPU (Unidade de Processamento de Tensor).
